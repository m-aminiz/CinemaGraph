```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_excel(r'E:\learning\dataVisulization\homeworks\hw7\movies.xlsx')
actor = pd.read_excel(r'E:\learning\dataVisulization\homeworks\hw7\actor.xlsx')
director = pd.read_excel(r'E:\learning\dataVisulization\homeworks\hw7\directors.xlsx')
metadata = pd.read_csv(r'E:\learning\dataVisulization\homeworks\hw7\movie_metadata.csv')
dir_mov = pd.read_excel(r'E:\learning\dataVisulization\homeworks\hw7\DirMov.xlsx')
act_mov = pd.read_excel(r'E:\learning\dataVisulization\homeworks\hw7\ActMov.xlsx')
```

```python
actor = actor.drop_duplicates()
actor.to_excel('nactor.xlsx')

director = director.drop_duplicates()
director.to_excel('ndirector.xlsx')
```

```python
def correction(genre):
    output = genre.split('|')
    return output[0]
          
movies['genres'] = movies['genres'].apply(correction) 
movies
# movies.to_excel('movies.xlsx')
```

```python
for id_ac,row_act in actor.iterrows():
    for id_meta,row_mata in metadata.iterrows():
        if row_act['actor_name'] == row_mata['actor_1_name']:
            act_mov = act_mov.append({'actor_id': id_ac,'movie_id': row_mata['id']}, ignore_index=True)

act_mov
```

```python
act_mov.to_excel('act_mov.xlsx')
```

```python
for id_dir,row_dir in director.iterrows():
    for id_meta,row_mata in metadata.iterrows():
        if row_dir['director_name'] == row_mata['director_name']:
            dir_mov = dir_mov.append({'director_id': id_dir,'movie_id': row_mata['id']}, ignore_index=True)

dir_mov
```

```python
dir_mov.to_excel('dir_mov.xlsx')
```
